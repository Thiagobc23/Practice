```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

bra = pd.read_csv('bra.csv', dtype = {'b': np.int32}, index_col = 'Indicator Name')
bra.head()
```

```python
# display some of the fields
list(bra)[275:300]
```

```python
# Self Improvement and Employment 
myPlt = bra.plot(y = ['Share of youth not in education, employment or training, total (% of youth population)',
                      'Unemployment, total (% of total labor force) (modeled ILO estimate)'])
myPlt.locator_params(integer=True)
plt.xlabel('Year')
plt.ylabel('%')
plt.title('Population not in education and Unemployed')
plt.legend(loc = 6)
plt.show()
```

```python
# plot wholesale and consumer prices index
myPlt = bra.plot(y = ['Wholesale price index (2010 = 100)', 'Consumer price index (2010 = 100)'])
# this makes sure the years will be displayed as int
myPlt.locator_params(integer=True)
# title, labels, legend location and display
plt.title('Wholesale and Consumer Prices')
plt.xlabel('Year')
plt.ylabel('Index')
plt.legend(loc = 2)
plt.show()

# bar chart with a time-series of inflation 
bra['Inflation, consumer prices (annual %)'].plot(kind = 'bar')
# title, labels and display
plt.title('Inflation, consumer prices')
plt.xlabel('Year')
plt.ylabel('%')
plt.show()

# show inflation in 2003
print('inflation 2003:', str(bra['Inflation, consumer prices (annual %)'].loc[2003]), ' %')
# pd.DataFrame(price.values*(inf.values/100), columns=price.columns, index=price.index).plot()
```

```python
#Check Correlation of some fields

import seaborn as sb

c = bra[['Service exports (BoP, current US$)',
 'Goods exports (BoP, current US$)',
 'Ores and metals imports (% of merchandise imports)',
 'Manufactures imports (% of merchandise imports)',
 'Livestock production index (2004-2006 = 100)',
 'Food production index (2004-2006 = 100)',
 'Crop production index (2004-2006 = 100)',
 'Public private partnerships investment in water and sanitation (current US$)',
 'Public private partnerships investment in transport (current US$)',
 'Public private partnerships investment in energy (current US$)',
 'Investment in water and sanitation with private participation (current US$)',
 'Investment in transport with private participation (current US$)',
 'Investment in energy with private participation (current US$)']]

labels = {
 'Service exports (BoP, current US$)':'SE',
 'Goods exports (BoP, current US$)':'GE',
 'Ores and metals imports (% of merchandise imports)':'OnMI',
 'Manufactures imports (% of merchandise imports)':'MI',
 'Livestock production index (2004-2006 = 100)':'LPi',
 'Food production index (2004-2006 = 100)':'FPi',
 'Crop production index (2004-2006 = 100)':'CPi',
 'Public private partnerships investment in water and sanitation (current US$)':'PPPI WS',
 'Public private partnerships investment in transport (current US$)':'PPPI T',
 'Public private partnerships investment in energy (current US$)':'PPPI E',
 'Investment in water and sanitation with private participation (current US$)':'IWSwPP',
 'Investment in transport with private participation (current US$)':'ITwPP',
 'Investment in energy with private participation (current US$)':'IEwPP'
}

corr = c.corr(method='pearson')
corr = corr.rename(labels, axis=1)
corr.head()
```

```python
# plot a correlation with a heat map
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = 1
sb.heatmap(corr, mask=mask)
```

```python
#Get a closer look at the correlation between service exports and livestock production index

cSnap = corr[['LPi', 'FPi', 'CPi', 'PPPI T']].loc[['Service exports (BoP, current US$)',
 'Goods exports (BoP, current US$)',
 'Ores and metals imports (% of merchandise imports)',
 'Manufactures imports (% of merchandise imports)',]]

cmap = sb.diverging_palette(255, 133, l=60, n=7, center="dark", as_cmap=True)
sb.heatmap(cSnap, cmap=cmap)
```

```python
# Kendall correlation
corr = c.corr(method='kendall')
corr = corr.rename(labels, axis=1)
sb.heatmap(corr, mask=mask).set_title('Kendall Correlation')
```

```python
#spearman correlation
corr = c.corr(method='spearman')
corr = corr.rename(labels, axis=1)
(sb.heatmap(corr, mask=mask)).set_title('Spearman Correlation')
```

```python
# Linear Regression
x = np.array(c['Service exports (BoP, current US$)'])
y = np.array(c['Livestock production index (2004-2006 = 100)'])

# arguments are x, y and order
pf = np.polyfit(x, y, 1)
pf2 = np.polyfit(x, y, 2)
pf3 = np.polyfit(x, y, 3)

# print intercept and slope
print(pf, '\n', pf2, '\n', pf3, '\n')
```

```python
# define figure to display charts
fig = plt.figure(figsize=(8, 15))

# add chart for the first regression
ax1 = fig.add_subplot(311)
ax1.plot(x,y,'^')
ax1 = plt.plot(x, np.polyval(pf,x))
ax1 = plt.title('First-order')
# add chart for the second regression
ax2 = fig.add_subplot(312)
ax2.plot(x,y,'^')
ax2 = plt.plot(x, np.polyval(pf2,x))
ax2 = plt.title('Second-order')
# add chart for the third regression
ax3 = fig.add_subplot(313)
ax3.plot(x,y,'^')
ax3 = plt.plot(x, np.polyval(pf3,x))
ax3= plt.title('Third-order')
```
